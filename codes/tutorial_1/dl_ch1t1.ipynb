```python
def model(x, w, b):
    return w * x + b
```

```python
def loss_fn(yy, y):
    squared_diffs = (yy - y)**2
    return squared_diffs.mean()
```

```python
w = torch.ones(())
b = torch.zeros(())
w, b
```

```python
yy = model(x, w, b)
yy
```

```python
loss = loss_fn(yy, y)
loss
```

```python
delta = 0.1
loss_rate_of_change_w = (loss_fn(model(x, w + delta, b), y) - loss_fn(model(x, w - delta, b), y)) / (2.0 * delta)
loss_rate_of_change_w
```
